# Create dataframe of postal code
Using BeautifulSoup to bring in the HTML

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

#open the URL and send to beautifulSoup
wikipagehtml_BS = BeautifulSoup(urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'))
wikipagehtml_BS
```

```python
import pandas as pd
```

```python
#scrap BeautifulSoup, too many steps. Can import the table directly into a pandas DF
df_original=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
df_original
```

```python
#checking postal codes to ensure correct
df_original['Postal Code'].unique() 
```

```python
#Two different methods to drop the Not Assigned status
#df_droppedassigned = df_original.query("Borough != 'Not assigned'")
df_droppedassigned = df_original[df_original.Borough != 'Not assigned'] 
```

```python
#Display all the rows in the table
pd.set_option('display.max_rows', None)
display(df_droppedassigned)
```

```python
# Display size of dataframe
print('The number of rows in my data frame is', df_droppedassigned.shape[0], 'rows')
```

```python
#Taken directly from lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto, Ontario are {}, {}.'.format(latitude, longitude))
```

```python
#conda install -c conda-forge folium
```

```python
#again copying from the lab
import folium # map rendering library
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(pd_BoroughLatLong['Latitude'], pd_BoroughLatLong['Longitude'], pd_BoroughLatLong['Borough'], pd_BoroughLatLong['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
